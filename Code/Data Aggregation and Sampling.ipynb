```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
path = r'C:\Users\tlhqu\Documents\GitHub\Data-visualization\Data\obama.csv'
df = pd.read_csv(path,parse_dates = ['year_month'])
df
```

```python
plt.plot(df.year_month, df.approve_percent,'o',markersize = 2, alpha= 0.3)
plt.show()
```

```python
df_mean = df.groupby('year_month').mean()
df_median = df.groupby('year_month').median()
```

```python
plt.plot(df_mean.index, df_mean.approve_percent,'red')
plt.plot(df_median.index, df_mean.approve_percent,'green')
plt.legend(['Mean','Median'])
plt.plot(df.year_month, df.approve_percent,'o',markersize = 2, alpha= 0.3)
plt.show()
```

```python
df_25 = df.groupby('year_month').quantile(0.25)
df_75 = df.groupby('year_month').quantile(0.75)
plt.plot(df_25.index, df_25.approve_percent,'green')
plt.plot(df_median.index, df_mean.approve_percent,'red')
plt.plot(df_75.index, df_75.approve_percent,'blue')
plt.plot(df.year_month, df.approve_percent,'o',markersize = 2, alpha= 0.3)
plt.legend(['25th Percentile','Median','75th Percentile'])
plt.show()
```
