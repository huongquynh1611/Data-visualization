```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = r'C:\Users\tlhqu\Documents\GitHub\Data-visualization\Data\countries.csv'
df = pd.read_csv(path)
df
```

```python
df_2007 = df[df.year == 2007]
df_2007
```

```python
plt.scatter(df_2007.gdpPerCapita, df_2007.lifeExpectancy, 5)
plt.title('GDP Per Capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Life Expectancy')
plt.show
```

```python
df_2007.gdpPerCapita.corr(df_2007.lifeExpectancy)
```

```python
plt.scatter(np.log10(df_2007.gdpPerCapita), df_2007.lifeExpectancy)
plt.title('GDP Per Capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Life Expectancy')
plt.show
```

```python
np.log10(df_2007.gdpPerCapita).corr(df_2007.lifeExpectancy)
```

```python
year_sorted = sorted(set(df.year))
```

```python
for given_year in year_sorted:
    data_year = df[df.year == given_year]
    plt.scatter(data_year.gdpPerCapita,data_year.lifeExpectancy,5)
    plt.title(given_year)
    plt.xlim(0,60000)
    plt.ylim(25,85)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
#     plt.show()
    plt.savefig(str(given_year), dpi = 200) # dpi is dots per inch
    plt.clf() # clear the curent plot
```

```python
'''
Examine Relationship Between GDP( Not per capita) and Life Expectancy in 2007
--> GDP = dgpPerCapita * population
Use both raw GDP and log of GDP
'''
```

```python
plt.scatter(df_2007.gdpPerCapita * df_2007.population, df_2007.lifeExpectancy, 5)
plt.title('GDP and Life Expectancy in 2007')
plt.xlabel('GDP ($)')
plt.ylabel('Life Expectancy')
plt.show
```

```python
(df_2007.gdpPerCapita * df_2007.population).corr(df_2007.lifeExpectancy)
```

```python
np.log10((df_2007.gdpPerCapita * df_2007.population)).corr(df_2007.lifeExpectancy)
```

```python
for given_year in year_sorted:
    data_year = df[df.year == given_year]
    plt.scatter(data_year.gdpPerCapita*data_year.population,data_year.lifeExpectancy,5)
    plt.title(given_year)
#     plt.xlim(0,2)
#     plt.ylim(25,80)
    plt.xlabel('GDP')
    plt.ylabel('Life Expectancy')
    plt.show()
```
